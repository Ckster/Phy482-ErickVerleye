```python
import matplotlib.pyplot as plt
%matplotlib inline

MagV=[]
R = 1
W=[]
L=1
dw=.01
w=0
while w <= 50:
    Magg = R/(((w*L)**2 + R**2)**(1/2))
    MagV.append(Magg)
    W.append(w)
    w+=dw
plt.plot(W,MagV)
plt.xlabel('Frequency')
plt.ylabel('Magnitude of V Ratio')
```

```python
import numpy as np
w=3
Vmaster=[]
T=[]
dt=.01
t=0
while t <100:
    V=[]
    for k in range(1,200):
        v = (4/np.pi)*(np.sin((2*k-1)*w*t))/((2*k-1))
        V.append(v)
    Vmaster.append(sum(V))
    T.append(t)
    t+=dt
plt.plot(T,Vmaster)
plt.xlabel('Time')
plt.ylabel('V In')
```
