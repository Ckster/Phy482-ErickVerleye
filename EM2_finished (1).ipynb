```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
import numpy as np
```

```python
x=np.arange(-10,10,.5)
y=np.arange(-10,10,.5)
X,Y=np.meshgrid(x,y)
```

```python
k=9*10**(9)
q=1
r2=X**2+Y**2
cos=X/(X**2+Y**2)**(1/2)
sin=Y/(X**2+Y**2)**(1/2)
Ex=(k*q*cos)/r2
Ey=(k*q*sin)/r2


```

```python
plt.figure(figsize=(8,8))
plt.quiver(X,Y,Ex,Ey)
#plt.xlim(-5,5)
#plt.ylim(-5,5)
```

```python
a=.5
rdip2p=(Y-4)**2+X**2
rdip2m=(Y+4)**2+X**2
EX = k*q*X/(rdip2p)**(3/2) - k*q*X/(rdip2m)**(3/2)
EY = k*q*(Y-4)/(rdip2p)**(3/2) - k*q*(Y+4)/(rdip2m)**(3/2)
```

```python
plt.figure(figsize=(8,8))
plt.quiver(X,Y,EX,EY)
#plt.xlim(-5,5)
#plt.ylim(-5,5)
```

```python
r1=(Y-3)**2+(X+3)**2
r2=(Y+3)**2+(X+3)**2
r3=(Y+3)**2+(X-3)**2
r4=(Y-3)**2+(X-3)**2
EXQ = k*q*(X+3)/(r1)**(3/2) - k*q*(X+3)/(r2)**(3/2) + k*q*(X-3)/(r3)**(3/2) - k*q*(X-3)/(r4)**(3/2)
EYQ = k*q*(Y-3)/(r1)**(3/2) - k*q*(Y+3)/(r2)**(3/2) + k*q*(Y+3)/(r3)**(3/2) - k*q*(Y-3)/(r4)**(3/2)
```

```python
plt.figure(figsize=(8,8))
plt.quiver(X,Y,EXQ,EYQ)
#plt.xlim(-5,5)
#plt.ylim(-5,5)
```
